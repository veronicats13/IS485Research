```python
import networkx as nx
import ModelConfig as mc
# import IndependentCascadesModelP001 as icm
# import IndependentCascadesModelP010 as icm
# import WeightedCascadeModel as icm
import ThresholdModel as tm
import operator
import random 
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
def GreedySelect(g, activated_nodes, model, config, num_iterations=100):
    ''' 
    Select node with max average marginal gain
    Variable 'mg' refers to 'marginal gain'
    Number of iterations to find average marginal gain defaulted to 100
    '''
    
    mg_dict = {}
    for candidate in g.nodes():
        if candidate in activated_nodes:
            continue

        mg = 0
        for i in range(num_iterations):
            config.add_model_initial_configuration("Infected", [candidate])
            model.set_initial_status(config)
            iterations = model.iteration_bunch()
            active_set_size, newly_activated_nodes = iterations
            newly_activated_nodes.append(candidate)
            mg += len(newly_activated_nodes)
            
            # print(f"Active Set Size: {active_set_size}")
            # print(f"Newly Activated Nodes: {newly_activated_nodes}")
            # print(f"Marginal Gain: {mg}")

            model.mg_reset(newly_activated_nodes)
        avg_mg = mg/num_iterations
        mg_dict[candidate] = avg_mg
    
    influencer = max(mg_dict.items(), key=operator.itemgetter(1))[0]
    print(f"Selected Influencer: Node {influencer}")
    print("")
    
    return influencer
```

```python
def RandomSelect(g, activated_nodes, model, config, num_iterations=100):
    ''' 
    Select Node with max influence spread in Random Sample
    Out of 100 attempts (iterations) of influence spread
    And 100 Random Samples
    '''
    influence_dict = {}
        
    sample_size = 5
    num_samples = 1

    # 100 samples
    for i in range(num_samples): 
        
        sample = []
        while len(sample) < sample_size:
            node = random.randint(0,len(g)-1)
            if node not in activated_nodes and node not in sample:
                sample.append(node)
        
        for node in sample:
            influence = 1
            influence_list = []
            for i in range(num_iterations):
                config.add_model_initial_configuration("Infected", [node])
                model.set_initial_status(config)
                iterations = model.iteration_bunch()
                active_set_size, newly_activated_nodes = iterations
                newly_activated_nodes.append(node)
                influence = len(activated_nodes) + len(newly_activated_nodes)

                # print(f"Active Set Size: {active_set_size}")
                # print(f"Newly Activated Nodes: {newly_activated_nodes}")
                # print(f"Influence: {influence}")

                model.mg_reset(newly_activated_nodes)
                influence_list.append(influence)
            influence_dict[node] = max(influence_list)
    
    influencer = max(influence_dict.items(), key=operator.itemgetter(1))[0]
    print(influence_dict)
    print(f"Selected Influencer: Node {influencer}")
    print("")
    
    return influencer
```

```python
def CentralitySort(g):
    '''
    Returns dictionary storing the average shortest distance of each node 
    '''
    asd_dict = {} 

    for node in g.nodes():
        count_dist = 0
        node_paths = nx.shortest_path(g)[node]
        for value in node_paths.values():
            count_dist += len(value) - 1 # excluding start node
        asd_dict[node] = count_dist/len(g)
    
    sorted_g = dict(sorted(asd_dict.items(), key=lambda item: item[1]))
    return sorted_g

def CentralitySelect(sorted_g, activated_nodes):
    ''' 
    Select node with highest shortest average distance
    Node cannot be previously activated
    '''
    
    for pair in sorted_g.items():
        candidate = pair[0]
        if candidate in activated_nodes:
            continue
        influencer = candidate
        
        print(f"Activated Nodes: {activated_nodes}")
        print(f"Influencer: Node {influencer}")
        print(f"Node's ASP: {pair[1]}")
        print("")
        
        return influencer

def remove_isolated_nodes(g):
    isolated_nodes = []
    for pair in g.degree():
        node = pair[0]
        degree = pair[1]
        if degree == 0:
            isolated_nodes.append(node)

    for node in isolated_nodes:
        g.remove_node(node)
        
    return g
```

```python
def RandomGreedySelect(g, activated_nodes, target_set_size, model, config, e=0.5, num_iterations=100):
    ''' 
    Select node with max average marginal gain in random sample
    Variable 'mg' refers to 'marginal gain'
    Number of iterations to find average marginal gain defaulted to 100
    '''
    n = len(g)
    k = target_set_size
    sample_size = math.ceil((n/k)*(math.log10(1/e))) # sample size
    print(f"n: {n}" )
    print(f"k: {k}" )
    print(f"log(1/e): {math.log10(1/e)}" )
    print(f"Sample Size: {sample_size}" )
    
    sample = []
    while len(sample) < sample_size:
        candidate = random.randint(0,len(g)-1)
        if candidate not in activated_nodes and candidate not in sample:
            sample.append(candidate)
    print(f"Sample: {sample}" )
    print(f"Activated Nodes: {activated_nodes}" )
       
    mg_dict = {}
    temp_activated_nodes = []
    for candidate in sample:
        
        mg = 0
        for i in range(num_iterations):
            config.add_model_initial_configuration("Infected", [candidate])
            model.set_initial_status(config)
            iterations = model.iteration_bunch()
            active_set_size, newly_activated_nodes = iterations
            newly_activated_nodes.append(candidate)
            mg += len(newly_activated_nodes)
            
            # print(f"Active Set Size: {active_set_size}")
            # print(f"Newly Activated Nodes: {newly_activated_nodes}")
            # print(f"Marginal Gain: {mg}")

            model.mg_reset(newly_activated_nodes)
        avg_mg = mg/num_iterations
        mg_dict[candidate] = avg_mg
    
    influencer = max(mg_dict.items(), key=operator.itemgetter(1))[0]
    
    print(mg_dict)
    print(f"Selected Influencer: Node {influencer}")
    print("")
    
    return influencer
```

```python
def InfluenceSpread(g, target_set_size, e=0.5, num_iterations=100, strategy='greedy'):
    '''
    Assuming continuous process (previously activated nodes cannot reattempt)
    As compared to one-time influence spread, all influencer same start point
    Default strategy set as greedy 
    '''
    # prepare graph for respective strategies
    g = remove_isolated_nodes(g)
    
    if strategy == 'centrality':
        sorted_g = CentralitySort(g)
    elif strategy == 'highdegree':
        sorted_g = sorted(g.degree, key=lambda x: x[1], reverse=True)
        
    # initialize diffusion model
    model = tm.ThresholdModel(g) 
    config = mc.Configuration()
    
    edge_dict = {}
    for (v, w) in list(g.edges()):
        counter = list(g.edges()).count((v, w))
        if (w, v) in list(g.edges()): 
            counter += list(g.edges()).count((w, v))
        edge_dict[(v, w)] = counter / len(list(g.neighbors(w)))

    threshold_dict = {}
    for node in g.nodes():
        threshold_dict[node] = np.random.random_sample()
    
    config.add_node_set_configuration('threshold', threshold_dict)
    config.add_edge_set_configuration('weight', edge_dict)
    
    # influence spread process
    all_activated_nodes = []
    active_ss_list = []
    # influencers_list = [] # variable currently not used
    
    for i in range(1, target_set_size+1):
        print(f"Selecting Influencer {i}")
        print("------------------------------------")
        
        if strategy == 'greedy':
            influencer = GreedySelect(g, all_activated_nodes, model, config, num_iterations)
        elif strategy == 'highdegree':
            influencer = HighDegreeSelect(g, sorted_g, all_activated_nodes)
        elif strategy == 'random':
            influencer = RandomSelect(g, all_activated_nodes, model, config, num_iterations)
        elif strategy == 'centrality':
            influencer = CentralitySelect(sorted_g, all_activated_nodes)
        elif strategy == 'randomgreedy':
            influencer = RandomGreedySelect(g, all_activated_nodes, target_set_size, model, config, e, num_iterations)
        
        # influencers_list.append(influencer)

        config.add_model_initial_configuration("Infected", [influencer]) #change "infected to activated"
        model.set_initial_status(config)
        iterations = model.iteration_bunch() 
        active_set_size, newly_activated_nodes = iterations

        newly_activated_nodes.append(influencer)
        all_activated_nodes.extend(newly_activated_nodes)
        
        print(f"Newly Activated Nodes: {newly_activated_nodes}")
        print(f"All Activated Nodes: {all_activated_nodes}")
        print(f"Active Set Size: {active_set_size}")
        print("")

        model.is_reset()
        active_ss_list.append(active_set_size)

    return active_ss_list
```

```python
def HighDegreeSelect(g, sorted_g, activated_nodes):
    ''' 
    Select node with highest degree/number of neighbours
    Node cannot be previously activated
    '''
    
    for pair in sorted_g:
        candidate = pair[0]
        if candidate in activated_nodes:
            continue
        influencer = candidate
        neighbors = list(g.neighbors(influencer))
        
        print(f"Activated Nodes: {activated_nodes}")
        print(f"Influencer: Node {influencer}")
        print(f"Degree: {pair[1]}")
        print(f"Neighbors: Nodes {neighbors}")
        print("")
        
        return influencer
```
