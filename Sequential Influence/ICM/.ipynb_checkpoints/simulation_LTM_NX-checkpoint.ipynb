```python
import numpy as np
import networkx as nx
import ModelConfig as mc
import ThresholdModel as tm
import operator
import random 
import math
import matplotlib.pyplot as plt
```

```python
def GreedySelect(g, activated_nodes, edge_dict, threshold_dict, num_iterations=100):
    ''' 
    Select node with max average marginal gain
    Variable 'mg' refers to 'marginal gain'
    Number of iterations to find average marginal gain defaulted to 100
    '''
  
    mg_dict = {}
    temp_activated_nodes = []
    for candidate in g.nodes():
        if candidate in activated_nodes:
            continue
        
        # clears the test node
        temp_activated_nodes.clear() 
        temp_activated_nodes.extend(activated_nodes)
        mg = 1 #count the influencer candidate node already

        for i in range(num_iterations):
            model = tm.ThresholdModel(g) 
            config = mc.Configuration()
            temp_activated_nodes.append(candidate)
            config.add_model_initial_configuration("Infected", temp_activated_nodes)
            config.add_node_set_configuration('threshold', threshold_dict)
            config.add_edge_set_configuration('weight', edge_dict)
            model.set_initial_status(config)
            iterations = model.iteration_bunch()

            for i in range(1,len(iterations)):
                for node in iterations[i]['status']:
                    mg += 1

        avg_mg = mg/num_iterations  
        mg_dict[candidate] = avg_mg
    
    influencer = max(mg_dict.items(), key=operator.itemgetter(1))[0]
    
    print(mg_dict)
    print(f"Selected Influencer: Node {influencer}")
    print("")
    
    return influencer
```

```python
def HighDegreeSelect(g, sorted_g, activated_nodes):
    ''' 
    Select node with highest degree/number of neighbours
    Node cannot be previously activated
    '''
    
    for pair in sorted_g:
        candidate = pair[0]
        if candidate in activated_nodes:
            continue
        influencer = candidate
        neighbors = list(g.neighbors(influencer))
        
        print(f"Activated Nodes: {activated_nodes}")
        print(f"Influencer: Node {influencer}")
        print(f"Degree: {pair[1]}")
        print(f"Neighbors: Nodes {neighbors}")
        print("")
        
        return influencer
```

```python
def RandomSelect(g, activated_nodes, sample_size, num_samples, edge_dict, threshold_dict, num_iterations=100):
    ''' 
    Select node with max average marginal gain
    Variable 'mg' refers to 'marginal gain'
    Number of iterations to find average marginal gain defaulted to 100
    '''
  
    influence_dict = {}
    temp_activated_nodes = []
        


    # 100 samples
    for i in range(num_samples): 
#         sample = np.random.choice(a = list(g.nodes()), size = sample_size, replace = False)
#         print('Original sample: ' + str(sample))
#         print(type(sample))

        # clears the test node
        temp_activated_nodes.clear() 
        temp_activated_nodes.extend(activated_nodes)
        
        # samples again if node is in activated nodes 
#         for node in sample: 
#             while node in activated_nodes: 
#                 idx = np.where(sample == node)
#                 np.delete(sample, idx)
#                 node = np.random.choice(a = list(g.nodes()), size = 1, replace = False)
#             np.append(sample, node)
#             print('Updated sample: ' + str(sample))
        sample = []
        while len(sample) < sample_size:
            node = random.randint(0,len(g)-1)
            if node not in activated_nodes and node not in sample:
                sample.append(node)
#         print(f"Sample: {sample}" )
#         print(f"Activated Nodes: {activated_nodes}" )
        
        for node in sample:
            influence = 1
            influence_list = []
            for i in range(num_iterations):
                model = tm.ThresholdModel(g) 
                config = mc.Configuration()
    #             for n in g.nodes():
    #                 if n in temp_activated_nodes:
    #                     config.add_node_configuration("attempt", n, 1)
    #                 else:
    #                     config.add_node_configuration("attempt", n, 0)
                temp_activated_nodes.append(node)
                config.add_model_initial_configuration("Infected", temp_activated_nodes)
                config.add_node_set_configuration('threshold', threshold_dict)
                config.add_edge_set_configuration('weight', edge_dict)
                model.set_initial_status(config)
                iterations = model.iteration_bunch()

                for i in range(1, len(iterations)):
                    for node in iterations[i]['status']:
                        influence += 1
                influence_list.append(influence)
            influence_dict[node] = max(influence_list)
    
    influencer = max(influence_dict.items(), key=operator.itemgetter(1))[0]
#     print('Updated sample: ' + str(sample))
    print(influence_dict)
    print(f"Selected Influencer: Node {influencer}")
    print("")
    
    return influencer
```

```python
def CentralitySort(g):
    '''
    Returns dictionary storing the average shortest distance of each node 
    '''
    asd_dict = {} 

    for node in g.nodes():
        count_dist = 0
        node_paths = nx.shortest_path(g)[node]
        for value in node_paths.values():
            count_dist += len(value) - 1 # excluding start node
        asd_dict[node] = count_dist/len(g)
    
    sorted_g = dict(sorted(asd_dict.items(), key=lambda item: item[1]))
    return sorted_g

def CentralitySelect(sorted_g, activated_nodes):
    ''' 
    Select node with highest shortest average distance
    Node cannot be previously activated
    '''
    
    for pair in sorted_g.items():
        candidate = pair[0]
        if candidate in activated_nodes:
            continue
        influencer = candidate
        
        print(f"Activated Nodes: {activated_nodes}")
        print(f"Influencer: Node {influencer}")
        print(f"Node's ASP: {pair[1]}")
        print("")
        
        return influencer
```

```python
def remove_isolated_nodes(g):
    isolated_nodes = []
    for pair in g.degree():
        node = pair[0]
        degree = pair[1]
        if degree == 0:
            isolated_nodes.append(node)

    for node in isolated_nodes:
        g.remove_node(node)
        
    return g
```

```python
def InfluenceSpread(g, target_set_size, e, num_iterations, strategy='greedy'):
    '''
    Assuming continuous process (previously activated nodes cannot reattempt)
    As compared to one-time influence spread, all influencer same start point
    Default strategy set as greedy 
    '''
    g = remove_isolated_nodes(g)
    
    if strategy == 'centrality':
        sorted_g = CentralitySort(g)
    elif strategy == 'highdegree':
        sorted_g = sorted(g.degree, key=lambda x: x[1], reverse=True)
        
#     if strategy == 'greedy' or strategy == 'randomgreedy' or strategy == 'random': 
    edge_dict = {}
    for (v, w) in list(g.edges()):
        counter = list(g.edges()).count((v, w))
        if (w, v) in list(g.edges()): 
            counter += list(g.edges()).count((w, v))
        edge_dict[(v, w)] = counter / len(list(g.neighbors(w)))

    threshold_dict = {}
    for node in g.nodes():
        threshold_dict[node] = np.random.random_sample()
    
    if strategy == 'random': 
        sample_size = 5
        num_samples = 100

    activated_nodes = []
    influencers_list = []
    active_ss_list = []
    for i in range(1, target_set_size+1):
        print(f"Selecting Influencer {i}")
        print("------------------------------------")
        
        if strategy == 'greedy':
            influencer = GreedySelect(g, activated_nodes, edge_dict, threshold_dict, num_iterations)
        elif strategy == 'highdegree':
            influencer = HighDegreeSelect(g, sorted_g, activated_nodes)
        elif strategy == 'random':
            influencer = RandomSelect(g, activated_nodes, sample_size, num_samples, edge_dict, threshold_dict, num_iterations)
        elif strategy == 'centrality':
            influencer = CentralitySelect(sorted_g, activated_nodes)
        elif strategy == 'randomgreedy': 
            influencer = RandomGreedySelect(g, activated_nodes, target_set_size, e, edge_dict, threshold_dict, num_iterations)
        influencers_list.append(influencer)
        model = tm.ThresholdModel(g) 
        config = mc.Configuration()
        
#         for n in g.nodes():
#             if n in activated_nodes:
#                 config.add_node_configuration("attempt", n, 1)
#             else:
#                 config.add_node_configuration("attempt", n, 0)
                
        activated_nodes.append(influencer)
        config.add_model_initial_configuration("Infected", activated_nodes)        
        config.add_node_set_configuration('threshold', threshold_dict)
        config.add_edge_set_configuration('weight', edge_dict)
        model.set_initial_status(config)
        iterations = model.iteration_bunch()
        
        for i in range(1,len(iterations)):
            for node in iterations[i]['status']:
                activated_nodes.append(node)
        active_ss_list.append(len(activated_nodes))

    return influencers_list, active_ss_list
```

```python
g = nx.erdos_renyi_graph(150, 0.1, directed=False)
target_set_size = 30
num_iterations = 100
e = 0.5
```

```python
strategy = 'highdegree'
_, hd_active_ss_list = InfluenceSpread(g, target_set_size, e, num_iterations, strategy)
hd_active_ss_list = np.array(hd_active_ss_list)
```

```python
strategy = 'centrality'
_, c_active_ss_list = InfluenceSpread(g, target_set_size, e, num_iterations, strategy)
c_active_ss_list = np.array(c_active_ss_list)
```

```python
strategy = 'randomgreedy'
num_iterations = 1
_, rg_active_ss_list = InfluenceSpread(g, target_set_size, e, num_iterations, strategy)
rg_active_ss_list = np.array(rg_active_ss_list)
```

```python
strategy = 'greedy'
_, g_active_ss_list = InfluenceSpread(g, target_set_size, e, num_iterations, strategy)
g_active_ss_list = np.array(g_active_ss_list)
```

```python
strategy = 'random'
_, r_active_ss_list = InfluenceSpread(g, target_set_size, e, num_iterations, strategy)
r_active_ss_list = np.array(r_active_ss_list)
```

```python
plt.plot(r_active_ss_list, label = "Random")
plt.plot(c_active_ss_list, label = "Centrality")
plt.plot(hd_active_ss_list, label = "High Degree")
plt.plot(g_active_ss_list, label = "Greedy")
plt.plot(rg_active_ss_list, label = "Random Greedy")

plt.xlabel('Target Set Size')
plt.ylabel('Active Set Size')
plt.title('Independent Cascade Model Pr0.10')

plt.legend()
plt.show()
```
