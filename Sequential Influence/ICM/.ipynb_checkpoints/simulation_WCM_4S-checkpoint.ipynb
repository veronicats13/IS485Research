```python
import networkx as nx
import ModelConfig as mc
# import IndependentCascadesModelP001 as icm
# import IndependentCascadesModelP010 as icm
import WeightedCascadeModel as icm
import operator
import random 
import matplotlib.pyplot as plt
import numpy as np
```

```python
def GreedySelect(g, activated_nodes, num_iterations=100):
    ''' 
    Select node with max average marginal gain
    Variable 'mg' refers to 'marginal gain'
    Number of iterations to find average marginal gain defaulted to 100
    '''
    
    mg_dict = {}
    temp_activated_nodes = []
    for candidate in g.nodes():
        if candidate in activated_nodes:
            continue
            
        temp_activated_nodes.clear()
        temp_activated_nodes.extend(activated_nodes)
        mg = 1 #count the influencer candidate node already

        for i in range(num_iterations):
            model = icm.IndependentCascadesModel(g) 
            config = mc.Configuration()
            for n in g.nodes():
                if n in temp_activated_nodes:
                    config.add_node_configuration("attempt", n, 1)
                else:
                    config.add_node_configuration("attempt", n, 0)
            temp_activated_nodes.append(candidate)
            config.add_model_initial_configuration("Infected", temp_activated_nodes)
            model.set_initial_status(config)
            iterations = model.iteration_bunch()

            for i in range(1,len(iterations)):
                for node in iterations[i]['status']:
                    mg += 1

        avg_mg = mg/num_iterations
        mg_dict[candidate] = avg_mg
    
    influencer = max(mg_dict.items(), key=operator.itemgetter(1))[0]
    
    print(mg_dict)
    print(f"Selected Influencer: Node {influencer}")
    print("")
    
    return influencer
```

```python
def HighDegreeSelect(g, sorted_g, activated_nodes):
    ''' 
    Select node with highest degree/number of neighbours
    Node cannot be previously activated
    '''
    
    for pair in sorted_g:
        candidate = pair[0]
        if candidate in activated_nodes:
            continue
        influencer = candidate
        neighbors = list(g.neighbors(influencer))
        
        print(f"Activated Nodes: {activated_nodes}")
        print(f"Influencer: Node {influencer}")
        print(f"Degree: {pair[1]}")
        print(f"Neighbors: Nodes {neighbors}")
        print("")
        
        return influencer
```

```python
def RandomSelect(g, activated_nodes):
    ''' 
    Select node with highest degree/number of neighbours
    Node cannot be previously activated
    '''
    
    selected = False
    while not selected:
        candidate = random.randint(0,len(g)-1)
        if candidate not in activated_nodes:
            influencer = candidate
            selected = True
    neighbors = list(g.neighbors(influencer))
    degree = len(neighbors)
        
    print(f"Activated Nodes: {activated_nodes}")
    print(f"Influencer: Node {influencer}")
    print(f"Degree: {degree}")
    print(f"Neighbors: Nodes {neighbors}")
    print("")
    
    return influencer
```

```python
def CentralitySort(g):
    '''
    Returns dictionary storing the average shortest distance of each node 
    '''
    asd_dict = {} 

    for node in g.nodes():
        count_dist = 0
        node_paths = nx.shortest_path(g)[node]
        for value in node_paths.values():
            count_dist += len(value) - 1 # excluding start node
        asd_dict[node] = count_dist/len(g)
    
    sorted_g = dict(sorted(asd_dict.items(), key=lambda item: item[1]))
    return sorted_g

def CentralitySelect(sorted_g, activated_nodes):
    ''' 
    Select node with highest shortest average distance
    Node cannot be previously activated
    '''
    
    for pair in sorted_g.items():
        candidate = pair[0]
        if candidate in activated_nodes:
            continue
        influencer = candidate
        
        print(f"Activated Nodes: {activated_nodes}")
        print(f"Influencer: Node {influencer}")
        print(f"Node's ASP: {pair[1]}")
        print("")
        
        return influencer
```

```python
def remove_isolated_nodes(g):
    isolated_nodes = []
    for pair in g.degree():
        node = pair[0]
        degree = pair[1]
        if degree == 0:
            isolated_nodes.append(node)

    for node in isolated_nodes:
        g.remove_node(node)
        
    return g
```

```python
def InfluenceSpread(g, target_set_size, num_iterations=100, strategy='greedy'):
    '''
    Assuming continuous process (previously activated nodes cannot reattempt)
    As compared to one-time influence spread, all influencer same start point
    Default strategy set as greedy 
    '''
    g = remove_isolated_nodes(g)
    
    if strategy == 'centrality':
        sorted_g = CentralitySort(g)
    elif strategy == 'highdegree':
        sorted_g = sorted(g.degree, key=lambda x: x[1], reverse=True)
    
    activated_nodes = []
    influencers_list = []
    for i in range(target_set_size):
        print(f"Selecting Influencer {i+1}")
        print("------------------------------------")
        
        if strategy == 'greedy':
            influencer = GreedySelect(g, activated_nodes, num_iterations)
        elif strategy == 'highdegree':
            influencer = HighDegreeSelect(g, sorted_g, activated_nodes)
        elif strategy == 'random':
            influencer = RandomSelect(g, activated_nodes)
        elif strategy == 'centrality':
            influencer = CentralitySelect(sorted_g, activated_nodes)
        influencers_list.append(influencer)
        model = icm.IndependentCascadesModel(g) 
        config = mc.Configuration()
        
        for n in g.nodes():
            if n in activated_nodes:
                config.add_node_configuration("attempt", n, 1)
            else:
                config.add_node_configuration("attempt", n, 0)
                
        activated_nodes.append(influencer)
        config.add_model_initial_configuration("Infected", activated_nodes)        
        model.set_initial_status(config)
        iterations = model.iteration_bunch()
        
        for i in range(1,len(iterations)):
            for node in iterations[i]['status']:
                activated_nodes.append(node)
                
    active_set_size = len(activated_nodes)
    return influencers_list, active_set_size
```

```python
g = nx.erdos_renyi_graph(150, 0.1) #number of nodes, probability to form edges

random_as_list = []
for i in range(30):
    target_set_size = i
    num_iterations = 100
    strategy = 'random'
    _, active_set_size = InfluenceSpread(g, target_set_size, num_iterations, strategy)
    random_as_list.append(active_set_size)
random_as_list = np.array(random_as_list)

centrality_as_list = []
for i in range(30):
    target_set_size = i
    num_iterations = 100
    strategy = 'centrality'
    _, active_set_size = InfluenceSpread(g, target_set_size, num_iterations, strategy)
    centrality_as_list.append(active_set_size)
centrality_as_list = np.array(centrality_as_list)

highdegree_as_list = []
for i in range(30):
    target_set_size = i
    num_iterations = 100
    strategy = 'highdegree'
    _, active_set_size = InfluenceSpread(g, target_set_size, num_iterations, strategy)
    highdegree_as_list.append(active_set_size)
highdegree_as_list = np.array(highdegree_as_list)

greedy_as_list = []
for i in range(5):
    target_set_size = i
    num_iterations = 1
    strategy = 'greedy'
    _, active_set_size = InfluenceSpread(g, target_set_size, num_iterations, strategy)
    greedy_as_list.append(active_set_size)
greedy_as_list = np.array(greedy_as_list)
```

```python
plt.plot(random_as_list, label = "Random")
plt.plot(centrality_as_list, label = "Centrality")
plt.plot(highdegree_as_list, label = "High Degree")
plt.plot(greedy_as_list, label = "Greedy")

plt.xlabel('Target Set Size')
plt.ylabel('Active Set Size')
plt.title('Weighted Cascade Model')

plt.legend()
plt.show()
```
